### Messages

```python
import os
from langchain_groq import ChatGroq
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
model=ChatGroq(model="qwen/qwen3-32b")
```

```python
model.invoke("Please tell me something about AI")
```

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages=[
    SystemMessage("You are a Data Sceientist Expert."),
    HumanMessage("Explain about RAG appplication")
]
response=model.invoke(messages)
print(response.content)
```

```python
system_msg=SystemMessage("You are a helpful coding assistant. Please provide the code with examples")

messages=[
    system_msg,
    HumanMessage("How do I Create a pyhton calculator")
]

response=model.invoke(messages)
print(response.content)
```

```python
### Detailed infor to LLM with System Message

system_msg=SystemMessage("""You are a helpful coding assistant.
 Always provide code examples and explain your reasoining.
 Be concise bu through your explainations""")

messages=[
    system_msg,
    HumanMessage("How do I Create a pyhton calculator")
]

response=model.invoke(messages)
print(response.content)
```
