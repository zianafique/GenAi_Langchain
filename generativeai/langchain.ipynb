### GenAI Application  Google Gemini AI Model

```python
from langchain.chat_models import init_chat_model
from langchain_openai import ChatOpenAI

model= init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
model
```

```python
response= model.invoke("How can I learn AI in 2026")
print(response.content)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
response= model.invoke("Write 100 words about Australia")
print(response.content)
```

### With GROQ Model

```python
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
model= init_chat_model("groq:llama-3.1-8b-instant")
response= model.invoke("Write 100 words about Australia")
print(response.content)
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="llama-3.1-8b-instant")
response= model.invoke("Why parrots Talk")
response
```

```python
response.content
```

### Streaming and Batch

```python
for chunk in model.stream("Write me a 1000 words about agentic ai? In one para"):
    print(chunk.text, end="", flush= True)
```

### Batch

```python
responses=model.batch(["Why do parrots talk", 
"How do airplane fly",
 "What is quantum computing?"])
for response in responses:
    print(response.content)
```

```python
responses=model.batch(["Why do parrots talk", 
"How do airplane fly",
 "What is quantum computing?"], config={'max_concurrency': 1 })
for response in responses:
    print(response.content)
```
